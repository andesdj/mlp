# Dummy Datasets 

* Crear listas, dataframe con información y distribuciones para pruebas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
n=1000
data=pd.DataFrame({
    'A'  :   np.random.randn(n),
    'B'  :   1.5+2.5*np.random.rand(n),
    'C'  :   np.random.uniform(5,32,n)
    })

data.describe()
```

```python
plt.hist(data['A'])
```

#  Agregacion de datos

```python
gender=["Male", 'Female']
income=['Poor', 'Middle', 'Rich']
```

```python
# Determina el tamaño del DataFrame
n=500

gender_data=[]
income_data=[]

for i in range(0,n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
# N (me, S)     -> m +s*Z            Z=DISTN
hmed=165
hdev=10

wmed=65
wdev=25
agemed=26
agedev=6

incomed=18000
incodev=3500
height=np.round(hmed+hdev*np.random.randn(n),2)
weight=np.round(wmed+wdev*np.random.randn(n),2)
age=np.ceil(agemed+agedev*np.random.randn(n))
age=age.astype(int)
income=np.round(incomed+incodev*np.random.randn(n),2)
```

```python
data=pd.DataFrame({
    'Gender': gender_data,
    'Ec. Status': income_data,
    'Height': height,
    'Weight':weight,
    'Age': age,
    'Income':income
    
})

data.describe()
```

## Agrupacion de datos al propio dataframe

* Crear un objeto de clase grupo con los elementos de un DataFrame
* EJ: grouped_gender.get_group('Female')
* grouped_gender.get_group('Female')

```python
grouped_gender=data.groupby('Gender')
grouped_gender.groups
```

### Iterar en los grupos e imprimir sus datos

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

#### Obtener la información de un grupo en especifico
 * Devuelve un DataFrame con la informacón del grupo

```python
grouped_gender.get_group('Female')
```

#### Agrupar con mas de un parámetro

```python
double_group=data.groupby(['Gender', 'Ec. Status'])
double_group
```

```python
for names, groups in double_group:
    print (names)
    print (groups)
```

#  Operaciones datos agrupados

* Agregar los datos despues de agrupar, ejecuta operaciones Ej Sumar los datos

```python
# DEvuelve el tamaño de cada uno de lso grupos
double_group.size()
```

```python
double_group.describe()
```

### Agregate para crear una agrgación personalizada


```python
double_group.aggregate(
{
    'Income':np.sum,
    'Age': np.mean,
    'Height': np.std  
})
```

```python
double_group.aggregate({
    'Age': np.mean,
    'Height': lambda h: np.mean(h)/np.std(h)
    
})
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x)/np.std(x)])
```

 # Filtrado de datos

```python
double_group['Age'].filter(lambda x: x.sum()>2400)
```

# Transformación de variables

```python
zscore= lambda x: (x-x.mean())/x.std()
z_group=double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
plt.hist(z_group['Age'])
```

```python
fill_na_mean=lambda x: x.fillna(x.mean())
double_group.transform(fill_na_mean)
```
