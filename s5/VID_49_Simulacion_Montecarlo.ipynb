## Simulación de MonteCarlo

* Generar 2 numeros Aleatorios x  e  entre 0 y 1
* calcular z= X^2 + Y^2    La formula del circulo
* Si z es Menor a 1 es dentro del circulo
* Si z no es menor que 1 esta fuera del circulo
* Calcular el numero total de veces que estan dentro del circulo y lo dividiremos entre el número total de intentos para obtener un aprox
* Usar esta probabilidad para aproximar al valor de PI y repetir el experimiento un Número suf para obtener diferentes aproximaciones A PI
* Calcular el promedio de n=1000 experimentos anteriores para obtener un valor final de PI


```python
# nexp : Numero de veces de experimentos que se lleva
# n: Numero total de veces  numero de puntos
# z: long vector en el circulo r=1
def pi_monte(n,nexp):
    n=n
    b=nexp
    pi_avg=0
    pi_value_list=[]
    for i in range(b):
        value=0
        x=np.random.uniform(0,1,n).tolist()
        y=np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z= np.sqrt(x[j]*x[j]+y[j]*y[j])
            if z<=1:
                value+=1
        float_value=float(value)
# pi_value: Valor de Pi ... Experimento * 4 
        pi_value =(float_value*4)/n
# pi_value_list: Lista con todos los valores de PI
        pi_value_list.append(pi_value)
# pi_avg: Valor acumulado en una variable con la suma de todos los valores de PI         
        pi_avg+=pi_value
# pi= Calculo de PI promedio
    pi=pi_avg/b
    print("ANDES")
    plt.plot(pi_value_list)
    plt.ylim([0, 5])
    return pi_value_list
    
```
