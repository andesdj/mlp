### Aprendizaje neuronal de señales de tráfico

```python
# Carga de las señales
```

```python
def load_data(data_directory):
    dirs= [ d for d in  os.listdir(data_directory)
            if os.path.isdir(os.path.join(data_directory, d))]
    labels=[]
    images=[]
    for d in dirs:
        label_dir= os.path.join(data_directory, d)
        file_names= [os.path.join(label_dir, f) 
                    for f in os.listdir(label_dir)
                    if f.endswith(".ppm")]
        for f in file_names:
            images.append(imd.imread(f))
            labels.append(int(d))
    
    return images, labels
    #print(label_dir)
    #print(file_names)
    #data=pd.read_csv('../src/datasets/ads/Advertising.csv')
```

```python
main_dir="../src/datasets/belgian/"
train_data_dir=os.path.join(main_dir,"Training")
test_data_dir=os.path.join(main_dir,"Testing")
images, labels= load_data(train_data_dir)
```

### Modelo de Red Neuronal con TensorFlow
##### Características
* Las imágenes no son del mismo tamaño (Se convierte a imagenes del mismo tamaño)
* Hay 62 clases de imágenes desde la 0 hasta la 61
* La distribución de señales de tráfico no es uniforme (algunas sales más veces que otras)
* TIP:  Generalmente el color no es esencial en las imágenes (Se converte a GrayScale)

```python
from skimage import transform
```

```python
w=999
h=999
for image in images:
    
    if image.shape[0]<h:
        h=image.shape[0]
    if image.shape[1]<w:
        w=image.shape[1]
print("Tamaños minimos (Alto x Ancho)  {0} x {1}".format(h,w))
```

### Apostar a una imagen con tamaño Minimo 30 30

```python

images30= [transform.resize(image, (30,30)) for image in images]
```

```python
rand_signs=random.sample(range(0,len(labels)), 6)
rand_signs
for i in range(len(rand_signs)):
    temp_im=images30[rand_signs[i]]
    plt.subplot(1,6,i+1)
    plt.axis("off")
    plt.imshow(temp_im)
    plt.subplots_adjust(wspace=0.5)
    plt.show()
    print("Forma:{0}, min:{1}, max{2}".format(temp_im.shape, temp_im.min(), temp_im.max()))
```

### Transformar a grayscale

```python
from skimage.color import rgb2gray
```

```python
images30 = np.array(images30)
images30 = rgb2gray(images30)
```

```python
rand_signs=random.sample(range(0,len(labels)), 6)
rand_signs
for i in range(len(rand_signs)):
    temp_im=images30[rand_signs[i]]
    plt.subplot(1,6,i+1)
    plt.axis("off")
    plt.imshow(temp_im, cmap="gray")
    plt.subplots_adjust(wspace=0.5)
    plt.show()
    print("Forma:{0}, min:{1}, max{2}".format(temp_im.shape, temp_im.min(), temp_im.max()))
```

```python
# 1 Construir placeholder
x= tf.placeholder(dtype=tf.float32, shape= [None, 30,30])
y=tf.placeholder(dtype= tf.int32, shape=[None])

# 2 Construir RNN
images_flat=tf.contrib.layers.flatten(x)
logits= tf.contrib.layers.fully_connected(images_flat,62, tf.nn.relu)

loss=tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels= y, logits=logits))

train_opt= tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)
final_pred= tf.argmax(logits,1)
accuracy= tf.reduce_mean(tf.cast(final_pred, tf.float32))
```

```python
# semilla de random
tf.set_random_seed(1234)
# Bucles de entrenamiento cada X iteracion EPOCH no conviene imprimir siempre el resultado de cada EPOC
sess= tf.Session()

sess.run(tf.global_variables_initializer())
# NUM ITERACIONES
n_epoch=601
n_iter=50
for i in range(n_epoch):
   
    _, accuracy_val= sess.run([train_opt, accuracy],
                             feed_dict={
                                 x:images30,
                                 y:list(labels)
                             })
   # _, loss_val= sess.run([train_opt, loss],
   #                          feed_dict={
   #                              x:images30,
   #                              y:list(labels)
   #                          })
    if i%n_iter==0:
        print ("EPOCH:  ", i)
        print("Eficacia ", accuracy_val )
    #    print("Loss: ", loss_val) 
        print ("   ")
    #print("Fin del EPOCH ", i)
    


```

### Evaluacion de la Red Neuronal

```python
#evaluar 16
sample_idx= random.sample(range (len (images30)),40)
sample_images= [images30[i] for i in sample_idx]
sample_labels= [labels[i] for i in sample_idx]
```

```python
sample_labels
```

```python
prediction= sess.run([final_pred], feed_dict={x:sample_images})[0]
prediction
```

```python
plt.figure(figsize=(16,20))
for i in range(len(sample_images)):
    truth= sample_labels[i]
    predi= prediction[i]
    plt.subplot(10,4, i+1)
    plt.axis("off")
    color="green" if truth==predi else "red"
    plt.text(32,15, "Real:      {0}\nPredict:  {1}".format(truth, predi) ,fontsize=14, color=color)
    plt.imshow(sample_images[i], cmap="gray")
plt.show()
```

```python
test_images, test_labels= load_data(test_data_dir)
test_images30=[transform.resize(im,( 30,30)) for im in test_images]
```

```python
len(test_images30)
```

```python
test_images30= rgb2gray(np.array(test_images30))
```

```python
prediction= sess.run([final_pred], feed_dict={x:test_images30})[0]
```

```python
match_count= sum([int(l0==lp) for l0, lp in zip(test_labels,prediction)])
match_count
```

```python
acc= match_count/len(test_labels)*100
print ("Eficacia de la red neuronal: {:.3f}%".format(acc))
```
