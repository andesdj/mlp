```python
import pandas as pd
import numpy as np
import os
%config IPCompleter.greedy=True
mpath='../src/datasets/ads'
pfile='Advertising.csv'
src=mpath+'/'+pfile
data=pd.read_csv(src, sep=',', header=0, )
```

```python
def co_pearson(df, var1,var2):
    df['corrn']=(df[var1]-np.mean(df[var1]))* (df[var2]-np.mean(df[var2]))
    df['corr1']= (df[var1]-np.mean(df[var1]))**2
    df['corr2']=(df[var2]-np.mean(df[var2]))**2
    corr_pearson=sum(df['corrn'])/np.sqrt(sum(df['corr1']*sum(df['corr2'])))
    return corr_pearson
```

```python
cols=data.columns.values
for x in cols:
    for y in cols:
        print (x+ "," +y + ":" + str(co_pearson(data, x,y)))
```
