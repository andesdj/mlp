## Regla de sturges: 
#### Define cuantos bloques deben obtenerse en un histograma para su correcta lectura 
+ Cuantas divisiones se hace el Histograma
* k=int(np.ceil(1+np.log2(Tamano Muestra)))

```python
# Obtener el largo de los campos para calcular con la regla de Sturges
data.describe

k=int(np.ceil(1+np.log2(3333)))

plt.hist(data['Day Calls'], bins=k)
plt.xlabel('NÃºmero de llamadas al dÃ­a')
plt.ylabel('Frecuencia')
plt.title('Histograma de nÃºmero de llamadas al dÃ­a')
```
