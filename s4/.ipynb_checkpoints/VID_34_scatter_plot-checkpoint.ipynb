```python
#MAgic plot en la misma celda
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
mpath='../src/datasets/customer-churn-model'
pfile='Customer Churn Model.txt'
src=mpath+'/'+pfile
data=pd.read_csv(src, sep=',', header=0, )
data.head()
```

```python
#Scatter plot

#savefig(path='RUTA DE GUARDAR IMAGEN.jpeg')
```

```python
data.plot(kind='scatter', x='Day Mins', y='Day Charge')
```

```python
data.plot(kind='scatter', x='Night Mins', y='Night Charge')
```

```python
figure,axs=plt.subplots(2,2,  sharey=True, sharex=True)
```

```python
figure,axs=plt.subplots(2,2,  sharey=True, sharex=True)
data.plot(kind='scatter', x='Day Mins', y='Day Charge', ax=axs[0][0])
data.plot(kind='scatter', x='Night Mins', y='Night Charge', ax=axs[0][1])
data.plot(kind='scatter', x='Day Calls', y='Day Charge', ax=axs[1][0])
data.plot(kind='scatter', x='Night Calls', y='Night Charge', ax=axs[1][1])
```

```python
# Obtener el largo de los campos para calcular con la regla de Sturges
data.describe
#Regla de sturges: Define cuantos bloques deben obtenerse en un histograma para su correcta lectura
k=int(np.ceil(1+np.log2(3333)))

plt.hist(data['Day Calls'], bins=k)
plt.xlabel('NÃºmero de llamadas al dÃ­a')
plt.ylabel('Frecuencia')
plt.title('Histograma de nÃºmero de llamadas al dÃ­a')
```
