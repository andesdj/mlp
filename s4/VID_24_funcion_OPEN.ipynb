# Cargar los datos desde la funcion OPEN y convertir a un nuevo formato

```python
import pandas as pd
import numpy as np
import os
#Abriri documentos con la funcion OPEN
mp='../src/datasets/customer-churn-model'
pfile='Customer Churn Model.txt'
src=mp+'/'+pfile
data1=open(src, 'r')
cols=data1.readline().strip().split(',')
# Contar la cantidad de columnas # columnas
ncols=len(cols)
ncols
#Para contar las filas se hara un Script 
#Contar las Filas del documento almacenado en col
counter=0
dic={}
#Crear diccionario de un archivo
# crear un for que recorrera las columnas y se pone clave:nombre columna, valor: [] Vacio
for col in cols:
    dic[col]=[]
# Se recorrera los valores de cada uno de las filas que será un array..y se seaparara el array que tiene coma (,).
for line in data1:
# Quita valores sobrantes y  almacenar el valor separado por comas en values
    values=line.strip().split(",")
# se hace una lista con la longitud de columnas  y se recorre ese tamaño
    for i in range(len(cols)):
# Se agrega a la columna i el valor del array en la posicion i        
        dic[cols[i]].append(values[i])
    counter+=1
print('EL data set tiene  %d filas y %d Columnas' %(counter, ncols))
```
