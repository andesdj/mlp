```python
%config IPCompleter.greedy=True
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import warnings
from math import sqrt
from collections import Counter

```

```python
dataset= {
    'k': [[1,2],[2,3], [3,1]],
    'r': [[6,5], [7,7] , [8,6]]
    
}
new_p=[5,7]
```

```python
[[plt.scatter(ii[0], ii[1], s=50, color=i) for ii in dataset [i]] for i in dataset]
plt.scatter(new_p[0], new_p[1], s=100, color="b")
```

```python
def knn(data, predict, k=3, verbose=False):
    if len(data)>=k:
        warnings.warn("\n K Es un valor menor que el número total de elementos a votar!!! ")
    distances=[]
    for group in data:
        for feature in data[group]:
            #d=sqrt((feature[0]-predict[0])**2+ (feature[1]-predict[1])**2)
            # Versión de Numpy de la misma forma pero optimizada
            #d= np.sqrt(np.sum((np.array(feature)-np.array(predict))**2))
            # Numpy y algebra lineal
            d= np.linalg.norm(np.array(feature)-np.array(predict))
            distances.append([d, group])
    if verbose:
        print("DISTANCES")
        print(" ")
        print(distances)
    # Ordena por la primera columna y la pone en una lista ordenada con los K - Primeros
    
    votes=[i[1] for i in sorted(distances)[:k]]
    
    if verbose:
        print("")
        print("VOTES")
        print(votes)
        print("")
   
    vote_result= Counter(votes).most_common(1)[0][0]
    if verbose:
        print("FINAL RESULT")
        print(vote_result)
    
    return vote_result
```

```python
new_p=[3,5]
result= knn(dataset, new_p)
result
```

```python
[[plt.scatter(ii[0], ii[1], s=50, color=i) for ii in dataset [i]] for i in dataset]
plt.scatter(new_p[0], new_p[1], s=200, color=result)
```

### Aplicar el codigo KNN al Dataset de  cáncer

```python
import pandas as pd
mpath='../src/datasets/cancer'
pfile='breast-cancer-wisconsin.data.txt'
src=mpath+'/'+pfile
df=pd.read_csv(src, sep=',' , header=None)
df.columns=["Names","V1","V2","V2","V4","V5","V6","V7","V8","V9","class"]
df.replace("?",-999999, inplace=True)    
df.head()
```

```python
#df=df.drop(["Names"], axis=1)
X=df[["V1","V2","V2","V4","V5","V6","V7","V8","V9"]]
Y=df['class']
df.head()
```

```python
full_data=df.astype(float).values.tolist()
full_data
```

```python
import random
```

```python
random.shuffle(full_data)
```

```python
test_size=0.2
```

```python
train_set={2:[], 4:[]}
test_set={2:[], 4:[]}
```

```python
train_data=full_data[:-int(test_size*len(full_data))]
test_data=full_data[-int(test_size*len(full_data)):]

```

```python
for i in train_data:
    train_set[i[-1]].append(i[:-1])

for i in test_data:
    test_set[i[-1]].append(i[:-1])
```

```python
# Datos del training
correct=0
total=0
for group in test_set:
    for data in test_set[group]:
        # K= 5 es por defecto en KNN de Python
        vote= knn(train_set, data, k=5)
        if group==vote:
            correct+=1
        total+=1
print("Eficacia del KK: ", correct/total)
```
